```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
doc = [64083, 21015, 20641, 22525, 21826, 81253, 75395]
```

```python
df = pd.DataFrame()
```

```python
for i in range(len(doc)):
    website = r"https://lex.uz/docs/{}".format(doc[i])
    response = requests.get(website)
    soup = BeautifulSoup(response.content, "html.parser")
    versiya = []
    
    for ultag in soup.find_all('ul', {'class': 'dropdown-content lx_dropdown_content'}):
        for litag in ultag.find_all('li'):
            versiya.append(litag.text)
    soha = []
    for ultag in soup.find_all('div', {'class': 'INDEXES_ON_REF'}):
        for litag in ultag.find_all('span', {'class': 'iorVal'}):
            soha.append(litag.text)
        

    print(versiya , "g", len(versiya), "g", soha)
```
